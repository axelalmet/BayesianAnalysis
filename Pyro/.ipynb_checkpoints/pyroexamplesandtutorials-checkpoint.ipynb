## Functions in Pyro

We use torch to sample the distributions and for many of the functions. In general, the point of using something like Torch or Tensorflow or Theano is all of the fancy tensor-valued and math functions. Note that many Pyro functions are simply wrappers to these Torch functions.

`````python
loc = 0.0 # mean zero
scale = 1.0 # unit variance
normal = torch.distributions.Normal(loc, scale) # create a normal distribution object
x = normal.rsample() # draw a sample from N(0,1)
print("sample", x)
print("log prob", normal.log_prob(x)) # score the sample from N(0,1)
`````

Pyro is implemented in a similar way to PyMC3, that is you have stochastic functions from which you sample and you can create deteministic functions based on these samples. You can also wrap all of the code as functions you can call repeatedly.

In this simple example, we predict the temperature depending on whether or not the weather is sunny or cloudy. The chance of the weather being cloudy follows a Bernoulli $\mathrm{B}(0.3)$ distribution, while the temperature follows a Normal distribution $\mathrm{N}(\mu, \sigma^2)$, with different means and standard deviations, depending on whether or not it is sunny or cloudy.

`````python
def weather():
    cloudy = pyro.sample('cloudy', pyro.distributions.Bernoulli(0.3))
    cloudy = 'cloudy' if cloudy.item() == 1.0 else 'sunny'
    mean_temp = {'cloudy': 55.0, 'sunny': 75.0}[cloudy]
    scale_temp = {'cloudy': 10.0, 'sunny': 15.0}[cloudy]
    temp = pyro.sample('temp', pyro.distributions.Normal(mean_temp, scale_temp))
    return cloudy, temp.item()

for _ in range(3):
    print(weather())
`````

Even better, we can take these functions and build further off them. Here, we use the weather distribution to predict ice cream sales.

`````python
def ice_cream_sales():
    cloudy, temp = weather()
    expected_sales = 200. if cloudy == 'sunny' and temp > 80.0 else 50.
    ice_cream = pyro.sample('ice_cream', pyro.distributions.Normal(expected_sales, 10.0))
    return ice_cream
`````

As Pyro is embedded in Python, we can use concepts like recursive functions that are terminated according to a random distribution, which is a really nice feature to have (essentially a stopping event). For example, we can use this concept to define a Geometric distribution.

`````python
def geometric(p, t=None):
    if t is None:
        t = 0
    x = pyro.sample("x_{}".format(t), pyro.distributions.Bernoulli(p))
    if x.item() == 1:
        return 0
    else:
        return 1 + geometric(p, t + 1) # This is the Recursive bit

print(geometric(0.5))
`````

Furthermore, we can define functions that take in stochastic functions as input. Note that the following function takes a single argument (scale) and generates three random variables.

`````python
def normal_product(loc, scale):
    z1 = pyro.sample("z1", pyro.distributions.Normal(loc, scale))
    z2 = pyro.sample("z2", pyro.distributions.Normal(loc, scale))
    y = z1 * z2
    return y

def make_normal_normal():
    mu_latent = pyro.sample("mu_latent", pyro.distributions.Normal(0, 1))
    fn = lambda scale: normal_product(mu_latent, scale)
    return fn

print(make_normal_normal()(1.))
`````

## Simple model in Pyro 

`````python
# Import some extra libraries
import pyro.infer
import pyro.optim
import pyro.distributions as dist
`````

Let us consider a simple model of noisy measurements of weight from a scale. Here, we are trying to infer the weight from the guess. This is an example of a latent variable.

$$
\sf weight\ |\ \sf guess \sim \mathrm{N}(\sf guess, 1) \\
\sf measurement\ | \sf weight, \sf guess \sim \mathrm{N}(\sf weight, 0.75).
$$

`````python
def scale(guess):
    weight = pyro.sample("weight", dist.Normal(guess, 1.0))
    return pyro.sample("measurement", dist.Normal(weight, 0.75))
`````

The whole reason we go to the trouble of developing probabilistic programming languages is so that we can build generative models and condition them on observed data, allowing us to infer the latent variables with which we build the model. In fact, we can do this using Pyro-specific functions.

Consider the above model. Suppose we want to infer the actual weight given that $\sf guess = 8.5$, but now we observe that $\sf measurement = 9.5$. Like in PyMC3, we can easily build this in by conditioning on scale via ```pyro.scale```, using the ```data``` argument.

`````python
# Define the conditioned scale
conditioned_scale = pyro.condition(scale, data={'measurement':9.5})
`````

In this way, we can define conditioned distributions via functions of the observed data/guesses.

`````python
def deferred_conditioned_scale(measurement, guess):
    return pyro.condition(scale, data={'measurement':measurement})(guess)
`````

Or we can just call ```sample``` directly.

`````python
def scale_obs(guess):  # equivalent to conditioned_scale above
    weight = pyro.sample("weight", dist.Normal(guess, 1.))
     # here we condition on measurement == 9.5
    return pyro.sample("measurement", dist.Normal(weight, 0.75), obs=9.5)
`````

Another nice function is ```pyro.do```, i.e. the famous do operator used in Judea Pearl's framework of causal inference. So Pyro can potentially be used for causal inference.

All inference is actually done via ```pyro.infer``` and its functions, e.g. ```pyro.infer.SVI``` (Stochastic Variational Inference).

For this example we actually know the exact form of the posterior distribution.

`````python
def perfect_guide(guess):
    loc =(0.75**2 * guess + 9.5) / (1 + 0.75**2) # 9.14
    scale = np.sqrt(0.75**2/(1 + 0.75**2)) # 0.6
    return pyro.sample("weight", dist.Normal(loc, scale))
`````

In general, it's unlikely we can actually do this. And often we'll have more complex functions of the following form:

`````python
def intractable_scale(guess):
    weight = pyro.sample("weight", dist.Normal(guess, 1.0))
    return pyro.sample("measurement", dist.Normal(some_nonlinear_function(weight), 0.75))
`````

Rather, what we do is use ```pyro.param``` to specify a _family_ of guides index by named parameters and search for the member of that family that is the best approximation according to some loss function. This approach is called _variational inference_.

`````python
# Example of setting parameter families, i.e. the damned priors...
simple_param_store = {}
a = simple_param_store.setdefault("a", torch.randn(1))
`````

So for the gues case, we're just inferring the mean and scale, $a$ and $b$.

`````python
def scale_parametrized_guide(guess):
    a = pyro.param("a", torch.tensor(guess))
    b = pyro.param("b", torch.tensor(1.))
    return pyro.sample("weight", dist.Normal(a, torch.abs(b)))
`````

Note that we need to enforce that the parameter b is positive. We can actually enforce that by using constraints provided by Torch:

`````python
from torch.distributions import constraints

def scale_parametrized_guide_constrained(guess):
    a = pyro.param("a", torch.tensor(guess))
    b = pyro.param("b", torch.tensor(1.), constraint=constraints.positive)
    return pyro.sample("weight", dist.Normal(a, b).to_event(1))  # no more torch.abs
`````

## Introduction to Stochastic Variational Inference

As a quick shortcut for where to put the various parts of a model, i.e. in the ```model(*args, **kwargs)``` function:

1. Observations ⟺ ````pyro.sample```` with the ```obs``` argument
2. Latent random variables ⟺ ```pyro.sample```
3. Parameters ⟺ ```pyro.param```

Before we explain SVI, let's start with some notation and definitions. Denote the observations by $\mathbf{x}$, the latent variables by $\mathbf{z}$, and the model parameters by $\theta$. Then we really want to estimate the parameters that define the joint distribution:
$$
p_\theta(\mathbf{x}, \mathbf{z}) = p_\theta(\mathbf{x}|\mathbf{z})p_\theta(\mathbf{z}).
$$
To make sure things are well behaved, we make a few assumptions about the probability distributions that constitute $p_\theta(\mathbf{x}, \mathbf{z})$:
1. We can sample from each of the constituent distributions, denoted by $p_i$.
2. We can compute the pointwise log of the probability density function (pdf) $p_i$.
3. The functions $p_i$ are differentiable with respect to the variables, $\theta$.

In the end, we're looking for the values of $\theta$ that maximise the log-evidence:
$$
\theta_\max = \mathrm{argmax}_\theta\log p_\theta(\mathbf{x}),
$$
where the log-evidence $\log p_\theta(\mathbf{x})$ is given straightforwardly by the integral:
$$
\log p_\theta(\mathbf{x}) = \log\int p_\theta(\mathbf{x}, \mathbf{z})d\mathbf{z}.
$$
To be clear, finding $\theta_\max$ is hard. Even if $\theta$ is fixed, the above integral is often intractable. Moreover, maximising the log-evidence is in general a non-convex optimisation problem. Good fucking luck.

Moreover, we'd also like to find the posterior over the latent variables:
$$
p_{\theta_\max}(\mathbf{z}|\mathbf{x}) = \frac{p_{\theta_\max}(\mathbf{x}, \mathbf{z})}{\int p_{\theta_\max}(\mathbf{x}, \mathbf{z})d\mathbf{z}}.
$$
Again, the integral is the difficult part. The idea here is that Variational Inference gives us a way to both find $\theta_\max$ and approximate the posterior $p_{\theta_\max}(\mathbf{z}|\mathbf{x})$.

The way we approximate the posterior is through the use of a _variational distribution_, or _guide_, $q_\phi(\mathbf{z})$, where $\phi$ are the variational parameters. This is encoded through the ```guide()``` function. It has a number of similarities to the ```model()``` function, in that it encodes ```pyro.sample``` and ```pyro.params``` calls. However, it does not contain the observed data, as the guide needs to actually be a normalised distribution. To this end, Pyro requires that both ```model()``` and ```guide()``` take the same arguments.

Once we've defined the variational distribution, the problem reduces to optimising in $\theta-\phi$ space that moves us closer to the exact posterior. Which begs the question: what are we optimising over, i.e. what's the objective function?

Luckily, we have one! This function is referred to as the _evidence lower bound (ELBO)_. The ELBO, which is a function of both the model and guide parameters, $\theta$ and $\phi$, is defined as:
$$
\mathrm{ELBO} = \mathbb{E}_{q_\phi(\mathbf{z})}\left[\log p_\theta(\mathbf{x}, \mathbf{z}) - \log q_\phi(\mathbf{z})\right].
$$
Note that we're assuming that we can calculate the log probabilities _inside_ the expectation. As we assume that we can sample from the guide distribution, this expectation can be calculated using Monte Carlo methods. What makes the ELBO special is that, for all $\theta$ and $\phi$:
$$
\log p_\theta(\mathbf{x}) \ge \mathrm{ELBO}.
$$
Hence, if we maximise the ELBO, we'll maximise the log-evidence in turn! Even better, the gap between the ELBO and the log-evidence can be related to the guide and posterior as such:
$$
\log p_\theta(\mathbf{x}) - \mathrm{ELBO} = \mathrm{KL}(q_\phi(\mathbf{z})\|p_\theta(\mathbf{z}|\mathbf{x})),
$$
where $\mathrm{KL}$ is the Kullback-Leibler divergence, which measures the "distance" between two distributions (relates to entropy and whatnot). I should point out that this equality really comes from considering the KL divergence between the guide distribution and the posterior. This equality shows that, for fixed $\theta$, taking steps in $\phi$ space to increase the ELBO brings the guide distribution closer to the posterior. Which is fantastic news for us, barring any technicalities of the pathological kind.

### Example
We consider the following example where you're trying to determine the ''fairness'' of a coin, based on the number of heads and tails we observe. Suppose that we have reason to believe that, while the coin may have started off as truly fair, it has deviated to become a bit unfair as it's been banged up over the years. This is encoded in the prior. 

In this case, we collect the data as a list of ```torch.tensor``` objects. We're trying to estimate the fairness, $f$, which is the latent variable here.

This is a standard Beta/Binomial example (like a two-group comparison). So the model is straightforward to define. 

`````python
import math
import os
import torch
import torch.distributions.constraints as constraints
import pyro
from pyro.optim import Adam
from pyro.infer import SVI, Trace_ELBO
import pyro.distributions as dist

# this is for running the notebook in our testing framework
smoke_test = ('CI' in os.environ)
n_steps = 2 if smoke_test else 2000

assert pyro.__version__.startswith('1.6.0')

# clear the param store in case we're in a REPL
pyro.clear_param_store()

# create some data with 6 observed heads and 4 observed tails
data = []
for _ in range(6):
    data.append(torch.tensor(1.0))
for _ in range(4):
    data.append(torch.tensor(0.0))

def model(data):
    # define the hyperparameters that control the beta prior
    alpha0 = torch.tensor(10.0)
    beta0 = torch.tensor(10.0)
    # sample f from the beta prior
    f = pyro.sample("latent_fairness", dist.Beta(alpha0, beta0))
    # loop over the observed data
    for i in range(len(data)):
        # observe datapoint i using the bernoulli likelihood
        pyro.sample("obs_{}".format(i), dist.Bernoulli(f), obs=data[i])
`````

We now need to define the corresponding guide, i.e. the variational distribution for the latent variable, $f$. Given that $f$ is meant to be interpreted as a probability, we only require that $f \in [0, 1.0]$. Therefore, it makes a lot of sense to use another Beta distribution which is characterised by two trainable parameters, $\alpha_q$ and $\beta_q$. In fact, we know that the Beta distribution is the conjugate prior, so it's exactly the right choice to approximate the posterior.

Some good guidelines to remember for the guide function:
- Make sure that the names of the random variables line up exactly between the model and guide.
- Make sure that the functions model(data) and guide(data) take the same arguments.
- Make sure that the variational parameters are torch.tensors. The requires_grad flag is automatically set to True by pyro.param.
- Moreover, given this model, make sure to use constraints like constraint=constraints.positive to ensure that alpha_q and beta_q remain non-negative during optimization.

`````python
def guide(data):
    # register the two variational parameters with Pyro
    # - both parameters will have initial value 15.0.
    # - because we invoke constraints.positive, the optimizer
    # will take gradients on the unconstrained parameters
    # (which are related to the constrained parameters by a log)
    alpha_q = pyro.param("alpha_q", torch.tensor(15.0),
                         constraint=constraints.positive)
    beta_q = pyro.param("beta_q", torch.tensor(15.0),
                        constraint=constraints.positive)
    # sample latent_fairness from the distribution Beta(alpha_q, beta_q)
    pyro.sample("latent_fairness", dist.Beta(alpha_q, beta_q))
`````

`````python
# setup the optimizer
adam_params = {"lr": 0.0005, "betas": (0.90, 0.999)}
optimizer = Adam(adam_params)

# setup the inference algorithm
svi = SVI(model, guide, optimizer, loss=Trace_ELBO())

# do gradient steps
for step in range(n_steps):
    svi.step(data)
    if step % 100 == 0:
        print('.', end='')
`````

`````python
# grab the learned variational parameters
alpha_q = pyro.param("alpha_q").item()
beta_q = pyro.param("beta_q").item()

# here we use some facts about the beta distribution
# compute the inferred mean of the coin's fairness
inferred_mean = alpha_q / (alpha_q + beta_q)
# compute inferred standard deviation
factor = beta_q / (alpha_q * (1.0 + alpha_q + beta_q))
inferred_std = inferred_mean * math.sqrt(factor)

print("\nbased on the data and our prior belief, the fairness " +
      "of the coin is %.3f +- %.3f" % (inferred_mean, inferred_std))
`````

## SVI Part II: Scaling this inference to larger datasets

For a model with $N$ observations, running the model, guide, and evaluating the Evidence Lower Bound (ELBO) scales badly with $N$. So if we use the above method, which samples the observations sequentially, for a large number of observations, this would take ages. The way to circumvent this is to employ subsampling, which is valid when the observations are conditionally independent given the latent variables.

In this case, the ELBOW is approximated by
$$
    \sum^N_{i=1}\log p(\mathbf{x}_i|\mathbf{z}) \approx \frac{N}{M}\sum_{i \in I_M}\log p(\mathbf{x}_i|\mathbf{z}),
$$
where $I_M$ is a mini-batch of indices where $M < N$.

## Pyro for Bayesian Regression

We'll show how Pyro can be used to implement Bayesian linear regression. That is, suppose we have a dataset $\mathcal{D}$ of the form:
$$
\mathcal{D} = \{(X_i, y_i)\}, \qquad i = 1, 2, \dots, N
$$
The goal is to fit a function of the form
$$
\mathbf{y} = \mathbf{w}\mathbf{X} + \mathbf{b} + \pmb{\epsilon}
$$
where $\mathbf{w}$ and $\mathbf{b}$ are the learnable weights and bias parameters, respectively, and $\pmb{\epsilon}$ represents the observation noise.

`````python
from torch import nn
from pyro.nn import PyroModule

assert issubclass(PyroModule[nn.Linear], nn.Linear)
assert issubclass(PyroModule[nn.Linear], PyroModule)
`````

Let's run a first pass and separate African and non-African nations and their fits. We also include an interaction term to separately model the effect of ruggedness for African and non-African nations.

`````python
# Dataset: Add a feature to capture the interaction between "cont_africa" and "rugged"
df["cont_africa_x_rugged"] = df["cont_africa"] * df["rugged"]
data = torch.tensor(df[["cont_africa", "rugged", "cont_africa_x_rugged", "rgdppc_2000"]].values,
                        dtype=torch.float)
x_data, y_data = data[:, :-1], data[:, -1]

# Regression model
linear_reg_model = PyroModule[nn.Linear](3, 1)

# Define loss and optimize
loss_fn = torch.nn.MSELoss(reduction='sum')
optim = torch.optim.Adam(linear_reg_model.parameters(), lr=0.05)
num_iterations = 1500 if not smoke_test else 2

def train():
    # run the model forward on the data
    y_pred = linear_reg_model(x_data).squeeze(-1)
    # calculate the mse loss
    loss = loss_fn(y_pred, y_data)
    # initialize gradients to zero
    optim.zero_grad()
    # backpropagate
    loss.backward()
    # take a gradient step
    optim.step()
    return loss

for j in range(num_iterations):
    loss = train()
    if (j + 1) % 50 == 0:
        print("[iteration %04d] loss: %.4f" % (j + 1, loss.item()))


# Inspect learned parameters
print("Learned parameters:")
for name, param in linear_reg_model.named_parameters():
    print(name, param.data.numpy())
`````

`````python
from pyro.nn import PyroSample

class BayesianRegression(PyroModule):
    def __init__(self, in_features, out_features):
        super().__init__()
        self.linear = PyroModule[nn.Linear](in_features, out_features)
        self.linear.weight = PyroSample(dist.Normal(0., 1.).expand([out_features, in_features]).to_event(2))
        self.linear.bias = PyroSample(dist.Normal(0., 10.).expand([out_features]).to_event(1))

    def forward(self, x, y=None):
        sigma = pyro.sample("sigma", dist.Uniform(0., 10.))
        mean = self.linear(x).squeeze(-1)
        with pyro.plate("data", x.shape[0]):
            obs = pyro.sample("obs", dist.Normal(mean, sigma), obs=y)
        return mean
`````

`````python
from pyro.infer.autoguide import AutoDiagonalNormal

model = BayesianRegression(3, 1)
guide = AutoDiagonalNormal(model)
`````

`````python
from pyro.infer import SVI, Trace_ELBO


adam = pyro.optim.Adam({"lr": 0.03})
svi = SVI(model, guide, adam, loss=Trace_ELBO())
`````

`````python
from collections import defaultdict
from pyro import poutine
from pyro.poutine.util import prune_subsample_sites
import warnings


class Predict(torch.nn.Module):
    def __init__(self, model, guide):
        super().__init__()
        self.model = model
        self.guide = guide

    def forward(self, *args, **kwargs):
        samples = {}
        guide_trace = poutine.trace(self.guide).get_trace(*args, **kwargs)
        model_trace = poutine.trace(poutine.replay(self.model, guide_trace)).get_trace(*args, **kwargs)
        for site in prune_subsample_sites(model_trace).stochastic_nodes:
            samples[site] = model_trace.nodes[site]['value']
        return tuple(v for _, v in sorted(samples.items()))

predict_fn = Predict(model, guide)
predict_module = torch.jit.trace_module(predict_fn, {"forward": (x_data,)}, check_trace=False)
`````

## Causal Variational Autoencoder (VAE)

`````python
import argparse
import logging

import torch

import pyro
import pyro.distributions as dist
from pyro.contrib.cevae import CEVAE

logging.getLogger("pyro").setLevel(logging.DEBUG)
logging.getLogger("pyro").handlers[0].setLevel(logging.DEBUG)
`````

`````python
def generate_data(args):
    """
    This implements the generative process of [1], but using larger feature and
    latent spaces ([1] assumes ``feature_dim=1`` and ``latent_dim=5``).
    """
    z = dist.Bernoulli(0.5).sample([args.num_data])
    x = dist.Normal(z, 5 * z + 3 * (1 - z)).sample([args.feature_dim]).t()
    t = dist.Bernoulli(0.75 * z + 0.25 * (1 - z)).sample()
    y = dist.Bernoulli(logits=3 * (z + 2 * (2 * t - 2))).sample()

    # Compute true ite for evaluation (via Monte Carlo approximation).
    t0_t1 = torch.tensor([[0.], [1.]])
    y_t0, y_t1 = dist.Bernoulli(logits=3 * (z + 2 * (2 * t0_t1 - 2))).mean
    true_ite = y_t1 - y_t0
    return x, t, y, true_ite
`````

`````python
def main(args):
    if args.cuda:
        torch.set_default_tensor_type('torch.cuda.FloatTensor')

    # Generate synthetic data.
    pyro.set_rng_seed(args.seed)
    x_train, t_train, y_train, _ = generate_data(args)

    # Train.
    pyro.set_rng_seed(args.seed)
    pyro.clear_param_store()
    cevae = CEVAE(feature_dim=args.feature_dim,
                  latent_dim=args.latent_dim,
                  hidden_dim=args.hidden_dim,
                  num_layers=args.num_layers,
                  num_samples=10)
    cevae.fit(x_train, t_train, y_train,
              num_epochs=args.num_epochs,
              batch_size=args.batch_size,
              learning_rate=args.learning_rate,
              learning_rate_decay=args.learning_rate_decay,
              weight_decay=args.weight_decay)

    # Evaluate.
    x_test, t_test, y_test, true_ite = generate_data(args)
    true_ate = true_ite.mean()
    print("true ATE = {:0.3g}".format(true_ate.item()))
    naive_ate = y_test[t_test == 1].mean() - y_test[t_test == 0].mean()
    print("naive ATE = {:0.3g}".format(naive_ate))
    if args.jit:
        cevae = cevae.to_script_module()
    est_ite = cevae.ite(x_test)
    est_ate = est_ite.mean()
    print("estimated ATE = {:0.3g}".format(est_ate.item()))
`````
