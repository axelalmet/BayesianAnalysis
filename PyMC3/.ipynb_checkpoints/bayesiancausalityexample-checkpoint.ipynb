# Bayesian causality example

```python
num_samples = 2000 # Set the number of samples for inference algorithms
num_tune_steps = 1000 # Set the number of tuning steps

# Set the observed data values
K1_obs = 12
K2_obs = 30
N1_obs = 100
N2_obs = 100
```

```python
num_elems = int(( (K2_obs + 1) * (K2_obs + 2) ) / 2)

m1_vals_list = []
m2_vals_list = []

for m1 in range(K2_obs + 1):
    m1_vals_list.append(np.repeat(m1, (K2_obs - m1 + 1)))
    m2_vals_list.append([i for i in range(K2_obs - m1 + 1)])
    
# Flatten the lists
m1_vals = np.array([item for sublist in m1_vals_list for item in sublist])
m2_vals = np.array([item for sublist in m2_vals_list for item in sublist])
```

```python
# Define the likelihood
def K2_likelihood(u, v):

    # This is defined as using Theano's tensor functionality
    elems = tt.gammaln(N2_obs + 1)  - tt.gammaln(m1_vals + 1) - tt.gammaln(N2_obs - m1_vals + 1) \
                    + tt.gammaln(N2_obs - m1_vals + 1) - tt.gammaln(m2_vals + 1) - tt.gammaln(N2_obs - m1_vals - m2_vals + 1) \
                    + tt.gammaln(N2_obs - m1_vals - m2_vals + 1) - tt.gammaln(K2_obs - m1_vals - m2_vals + 1) - tt.gammaln(N2_obs - K2_obs + 1) \
                    + (K2_obs - m2_vals) * pm.math.log(u) \
                    + (N2_obs - K2_obs + m2_vals) * pm.math.log(1.0 - u) \
                    + (K2_obs - m1_vals) * pm.math.log(v) \
                    + (N2_obs - K2_obs + m1_vals) * pm.math.log(1.0 - v)
        
    # Take the log-sum-exp (with the max) to make sure we don't get overflow or underflow
    return tt.max(elems) + pm.logsumexp(elems - tt.max(elems))
```

```python
with pm.Model() as treatment_causality_model:

#     BoundedBeta = pm.Bound(pm.Beta, lower=0.0, upper=1.0)
    # Define the priors
    u = pm.Beta('u', alpha=1 + K1_obs, beta=1 + N1_obs - K1_obs)
    v = pm.Beta('v', alpha=1, beta=1)

    like = pm.Potential('K2', K2_likelihood(u=u, v=v))
                         
    recovery_due_to_treatment = pm.Deterministic('prob_v', v / (u + v - u * v))
    
    # Sample from the posterior
    trace = pm.sample(num_samples, tune=num_tune_steps, return_inferencedata=False)
    prior = pm.sample_prior_predictive()
    posterior_predictive = pm.sample_posterior_predictive(trace, var_names = ['u', 'v'])
    
pm.model_to_graphviz(treatment_causality_model)
```

```python
pm_data = az.from_pymc3(
        trace=trace,
        prior=prior,
        posterior_predictive=posterior_predictive)
```

# Effects of cause model in the style of Rubin

```python
# Simulate the dataset
N = 500 # Total sample size
alpha = 1.0 # Intercept in the y-model
tau = 0.25 # Treatment effect

# Assignment mechanism
N_t = 200
W = np.random.choice(2, N) # Randomly assign treatment

# Simulate the outcomes
mu_c = alpha # Mean for control
sd_c = 1.0 # Standard deviation for control
mu_t = alpha + tau # Mean for treatment
sd_t = 1.0 # Standard deviation for control
means = [mu_c, mu_t]

rho = 0.0 # Correlation
cov_mat = np.array([[sd_c**2.0, rho * sd_c * sd_t], [rho * sd_c * sd_t, sd_t**2.0]]) # Covariant matrix

outcomes = np.random.multivariate_normal(means, cov_mat, size=N) # Generate via multivariate normal
Y0 = outcomes[:, 0]        # potential outcome if W = 1
Y1 = outcomes[:, 1]        # potential outcome if W = 0
tau_unit = Y1 - Y0

# The realization of potential outcomes by the assignment mechanism
Y_obs = Y0 * (1.0 - W) + Y1 * W # Observed outcomes
Y_mis = Y0 * W + Y1 * (1.0 - W) # Missing outcomes
```

```python
# Now let's try and build this model in PyMC3
with pm.Model() as model:

    # Define the priors
    alpha = pm.Normal('alpha', mu=0.0, sigma=5.0)
    tau = pm.Normal('tau', mu=0.0, sigma=5.0)
    sigma_c = pm.Normal('sigma_c', mu=0.0, sigma=5.0)
    sigma_t = pm.Normal('sigma_t', mu=0.0, sigma=5.0)
    mu = alpha + tau * W
    sigma = sigma_t * W + sigma_c * (1.0 - W)
    
    y_obs = pm.Normal('Y_obs', mu=mu, sigma=sigma, observed=Y_obs)
    
    # Sample from the posterior
    trace = pm.sample(2000, tune=1000, start={'alpha': np.array(0.1), 'tau': np.array(0.1), 'sigma_c': np.array(0.1), 'sigma_t': np.array(0.1)})
```

```python
az.summary(trace)
```

```python
pm.plot_posterior(trace, var_names=['alpha', 'tau'])
```
