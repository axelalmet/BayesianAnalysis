In this notebook, we will reproduce the causal inference example on Ferenc Huszár's [blog](https://www.inference.vc/causal-inference-2-illustrating-interventions-in-a-toy-example/) using the PyMC3 code written by Fabio Zennaro [here](https://github.com/FMZennaro/CausalInference/blob/master/Causal%20Inference.ipynb). This code is purely for pedadogical purposes and to see how causal inference can be implemented using PyMC3. All the credit goes to Huszár and Zennaro for the example and code, respectively.

```python
# Import the relevant libraries
import pymc3 as pm
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
num_samples = 1000 # Set the number of samples for inference algorithms
```

We define some helper functions for plotting.

```python
def jointplot(x,y,color,title):
    g = sns.jointplot(x, y, color=color)
    g.annotate(stats.pearsonr)
    g.set_axis_labels(xlabel='x', ylabel='y')
    g.fig.suptitle(title)
    
def kdeplot(y,color,title):
    g = sns.kdeplot(y, color=color, label=title)
```

Let us now define the three observational models presented by Huszár, which are generated in very different ways but generate the same observational joint distribution $(X, Y)$. We define these models using PyMC3's functionality.

```python
def model1():
    with pm.Model():
        x = pm.Normal('x', mu=0, sd=1)
        y = pm.Deterministic('y',x + 1 + np.sqrt(3)*pm.Normal('n0', mu=0, sd=1))
        trace = pm.sample(num_samples)
        return trace['x'],trace['y']
    
def model2():
    with pm.Model():
        y = pm.Deterministic('y', 1 + 2*pm.Normal('n0', mu=0, sd=1))
        x = pm.Deterministic('x', (y-1)/4 + np.sqrt(3)*pm.Normal('n1', mu=0, sd=1)/2)
        trace = pm.sample(num_samples)
        return trace['x'],trace['y']
    
def model3():
    with pm.Model():
        z = pm.Normal('z', mu=0, sd=1)
        y = pm.Deterministic('y',z + 1 + np.sqrt(3)*pm.Normal('n0', mu=0, sd=1))
        x = pm.Deterministic('x',z)
        trace = pm.sample(num_samples)
        return trace['x'],trace['y']
```

As an exercise, we will also plot the directed acyclic graphs (DAGs) associated with each model1 using the [causalgraphicalmodels](https://github.com/ijmbarr/causalgraphicalmodels) package.

```python
from causalgraphicalmodels import CausalGraphicalModel
```

```python
model1_dag = CausalGraphicalModel(nodes=['X', 'Y'], edges=[('X', 'Y')])
model2_dag = CausalGraphicalModel(nodes=['X', 'Y'], edges=[('Y', 'X')])
model3_dag = CausalGraphicalModel(nodes=['X', 'Y', 'Z'], edges=[('Z', 'X'), ('Z', 'Y')])
```

```python
model1_dag.draw()
```

```python
model2_dag.draw()
```

```python
model3_dag.draw()
```

We use PyMC3 to sample and plot the distributions.

```python
x,y = model1()
jointplot(x, y, color='blue', title='Observational P(X,Y) for model1')
 
x,y = model2()
jointplot(x, y, color='green', title='Observational P(X,Y) for model2')
 
x,y = model3()
jointplot(x, y, color='red', title='Observational P(X,Y) for model3')
```

So the observable joint distributions, $(X, Y)$, are all the same. So how do we show they're different? By counterfactuals via do-calculus. That is, we can ask ourselves how the distributions will behave if we hard-code a conditional value, $X = 3$, say. Is the same as the observational conditional distribution? Not quite and we can show this. We have to redefine the PyMC3 models but it's straightforward to set the observed value, $X = 3$.

```python
def model1_observe_X_3():
    with pm.Model():
        x = pm.Normal('x', mu=0, sd=1,observed=3)
        y = pm.Deterministic('y',x + 1 + np.sqrt(3)*pm.Normal('n0', mu=0, sd=1))
        trace = pm.sample(num_samples)
        return trace['y']
    
def model2_observe_X_3():
    with pm.Model():
        y = pm.Deterministic('y', 1 + 2*pm.Normal('n0', mu=0, sd=1))
        x = pm.Normal('x', mu=(y-1)/4.0, sd=3/4.0, observed=3)
        trace = pm.sample(num_samples)
        return trace['y']
    
def model3_observe_X_3():
    with pm.Model():
        z = pm.Normal('z', mu=0, sd=1,observed=3)
        y = pm.Deterministic('y',z + 1 + np.sqrt(3)*pm.Normal('n0', mu=0, sd=1))
        x = pm.Deterministic('x',z)
        trace = pm.sample(num_samples)
        return trace['y']
```

```python
y1 = model1_observe_X_3()
y2 = model2_observe_X_3()
y3 = model3_observe_X_3()

plt.figure()
plt.title('Observational P(Y | X=3)')
kdeplot(y1, color='blue', title='model1')
kdeplot(y2, color='green', title='model2')
kdeplot(y3, color='red', title='model3')
plt.legend()
```

So the observational conditional distributions $Y|X=3$ seem to be roughly the same. Now we will show what happens when we directly force $X = 3$, i.e. the intervention. These distributions are defined as $P(Y|do(X=3))$.

In the language of causal graphs, interventions are performed by mutilating the graphs. That is, we remove all edges pointing towards the variable on which we intervene, i.e. $X$. For comparison, we show what the causal graphs look like now.

```python
model1_do_dag = CausalGraphicalModel(nodes=['X', 'Y'], edges=[('X', 'Y')])
model2_do_dag = CausalGraphicalModel(nodes=['X', 'Y'], edges=[])
model3_do_dag = CausalGraphicalModel(nodes=['X', 'Y', 'Z'], edges=[ ('Z', 'Y')])
```

```python
model1_do_dag.draw()
```

```python
model2_do_dag.draw()
```

```python
model3_do_dag.draw()
```

From here, we can see that it is only for model 1 where $P(Y|do(X)) = p(y|x)$, i.e. the interventional and conditional distributions coincide. For model 2 and 3, we can see that $P(Y|do(X)) = p(y)$.

```python
def model1_do_X_3():
    with pm.Model():
        x = 3
        y = pm.Deterministic('y',x + 1 + np.sqrt(3)*pm.Normal('n0', mu=0, sd=1))
        trace = pm.sample(num_samples)
        return trace['y']
    
def model2_do_X_3():
    with pm.Model():
        y = pm.Deterministic('y', 1 + 2*pm.Normal('n0', mu=0, sd=1))
        x = 3
        trace = pm.sample(num_samples)
        return trace['y']
    
def model3_do_X_3():
    with pm.Model():
        z = pm.Normal('z', mu=0, sd=1)
        y = pm.Deterministic('y',z + 1 + np.sqrt(3)*pm.Normal('n0', mu=0, sd=1))
        x = 3
        trace = pm.sample(num_samples)
        return trace['y']
```

```python
y1 = model1_do_X_3()
y2 = model2_do_X_3()
y3 = model3_do_X_3()

plt.figure()
plt.title('Interventional P(Y | do(X=3))')
kdeplot(y1, color='blue', title='model1')
kdeplot(y2, color='green', title='model2')
kdeplot(y3, color='red', title='model3')
plt.legend()
```

One of the nice aspects of do-calculus is that, assuming confounders aren't playing a role, we can actually infer the interventional model directly from the non-intervened model. Note that these are generated with the help of the mutilated causal graphs.

```python
def model1_docalculus_X_3():
    with pm.Model():
        x = pm.Normal('x', mu=0, sd=1,observed=3)
        y = pm.Deterministic('y',x + 1 + np.sqrt(3)*pm.Normal('n0', mu=0, sd=1))
        trace = pm.sample(num_samples)
        return trace['y']
    
def model2_docalculus_X_3():
    with pm.Model():
        y = pm.Deterministic('y', 1 + 2*pm.Normal('n0', mu=0, sd=1))
        trace = pm.sample(num_samples)
        return trace['y']
    
def model3_docalculus_X_3():
    with pm.Model():
        z = pm.Normal('z', mu=0, sd=1)
        y = pm.Deterministic('y',z + 1 + np.sqrt(3)*pm.Normal('n0', mu=0, sd=1))
        trace = pm.sample(num_samples)
        return trace['y']
```

```python
y1 = model1_docalculus_X_3()
y2 = model2_docalculus_X_3()
y3 = model3_docalculus_X_3()

plt.figure()
plt.title('Interventional P(Y | do(X=3)) evaluated from observational data via do-calculus')
kdeplot(y1, color='blue', title='model1')
kdeplot(y2, color='green', title='model2')
kdeplot(y3, color='red', title='model3')
plt.legend()
```

These results match the results from the direct interventional distribution, showing that we can calculate the interventional distributions by using the observed conditional distributions which have been transformed via do-calculus.

We can also inspect the individual counterfactual distributions, which are derived the same way, via do-calculus. We compare the counterfactuals to the factual models to show how they change under the interventions.

```python
def model1_counterfactual_X_3():
    with pm.Model():
        x = pm.Normal('x', mu=0, sd=1)
        y = pm.Deterministic('y',x + 1 + np.sqrt(3)*pm.Normal('n0', mu=0, sd=1))
        trace = pm.sample(1,chains=1)
        
        factual_x = trace['x']
        factual_y = trace['y']
        factual_n0 = trace['n0']
        
        counterfactual_x = 3
        counterfactual_y = counterfactual_x + 1 + factual_n0
        
        return factual_x, factual_y, counterfactual_x, counterfactual_y

def model2_counterfactual_X_3():
    with pm.Model():
        y = pm.Deterministic('y', 1 + 2*pm.Normal('n0', mu=0, sd=1))
        x = pm.Deterministic('x', (y-1)/4 + np.sqrt(3)*pm.Normal('n1', mu=0, sd=1)/2)
        trace = pm.sample(1,chains=1)
        
        factual_x = trace['x']
        factual_y = trace['y']
        factual_n0 = trace['n0']
        factual_n1 = trace['n1']

        counterfactual_x = 3
        counterfactual_y = factual_y
        
        return factual_x, factual_y, counterfactual_x, counterfactual_y

def model3_counterfactual_X_3():
    with pm.Model():
        z = pm.Normal('z', mu=0, sd=1)
        y = pm.Deterministic('y',z + 1 + np.sqrt(3)*pm.Normal('n0', mu=0, sd=1))
        x = pm.Deterministic('x',z)
        trace = pm.sample(1,chains=1)
        
        factual_z = trace['x']
        factual_x = trace['x']
        factual_y = trace['y']
        factual_n0 = trace['n0']

        counterfactual_x = 3
        counterfactual_y = factual_z + 1 + np.sqrt(3)*factual_n0
        
        return factual_x, factual_y, counterfactual_x, counterfactual_y
```

```python
factual_x, factual_y, counterfactual_x, counterfactual_y = model1_counterfactual_X_3()
print('MODEL1 - Factual (x,y): {0}, {1}'.format(factual_x, factual_y))
print('MODEL1 - Counterfactual (x,y): {0}, {1}'.format(counterfactual_x, counterfactual_y))

factual_x, factual_y, counterfactual_x, counterfactual_y = model1_counterfactual_X_3()
print('MODEL1 - Factual (x,y): {0}, {1}'.format(factual_x, factual_y))
print('MODEL1 - Counterfactual (x,y): {0}, {1}'.format(counterfactual_x, counterfactual_y))

factual_x, factual_y, counterfactual_x, counterfactual_y = model2_counterfactual_X_3()
print('MODEL2 - Factual (x,y): {0}, {1}'.format(factual_x, factual_y))
print('MODEL2 - Counterfactual (x,y): {0}, {1}'.format(counterfactual_x, counterfactual_y))

factual_x, factual_y, counterfactual_x, counterfactual_y = model2_counterfactual_X_3()
print('MODEL2 - Factual (x,y): {0}, {1}'.format(factual_x, factual_y))
print('MODEL2 - Counterfactual (x,y): {0}, {1}'.format(counterfactual_x, counterfactual_y))

factual_x, factual_y, counterfactual_x, counterfactual_y = model3_counterfactual_X_3()
print('MODEL3 - Factual (x,y): {0}, {1}'.format(factual_x, factual_y))
print('MODEL3 - Counterfactual (x,y): {0}, {1}'.format(counterfactual_x, counterfactual_y))

factual_x, factual_y, counterfactual_x, counterfactual_y = model3_counterfactual_X_3()
print('MODEL3 - Factual (x,y): {0}, {1}'.format(factual_x, factual_y))
print('MODEL3 - Counterfactual (x,y): {0}, {1}'.format(counterfactual_x, counterfactual_y))
```

As we expected, only the counterfactual values of $Y$ for model 1 change with respect to the factual values.

Finally, we can compute the counterfactual distributions $P(Y^*|X^*=3)$ much in the same way we compute the factual distributions via PyMC3. Note that we're really just combining the previous models together.

```python
def model1_counterfactual_X_3():
    with pm.Model():
        x = pm.Normal('x', mu=0, sd=1)
        y = pm.Deterministic('y',x + 1 + np.sqrt(3)*pm.Normal('n0', mu=0, sd=1))
        trace = pm.sample(2 * num_samples, chains=1)
        
        # Simulate the factual model
        factual_x = trace['x']
        factual_y = trace['y']
        factual_n0 = trace['n0']

        counterfactual_x = 3 * np.ones(2 * num_samples)
        counterfactual_y = counterfactual_x + 1 + factual_n0
        
        return factual_x, factual_y, counterfactual_x, counterfactual_y
    
def model2_counterfactual_X_3():
    with pm.Model():
        # Simulate the factual model
        y = pm.Deterministic('y', 1 + 2*pm.Normal('n0', mu=0, sd=1))
        x = pm.Deterministic('x', (y-1)/4 + np.sqrt(3)*pm.Normal('n1', mu=0, sd=1)/2)
        trace = pm.sample(2 * num_samples, chains=1)
        
        factual_x = trace['x']
        factual_y = trace['y']
        factual_n0 = trace['n0']

        counterfactual_x = 3 * np.ones(2 * num_samples)
        counterfactual_y = factual_y
        
        return factual_x, factual_y, counterfactual_x, counterfactual_y
    
def model3_counterfactual_X_3():
    with pm.Model():
        # Simulate the factual model
        z = pm.Normal('z', mu=0, sd=1)
        y = pm.Deterministic('y',z + 1 + np.sqrt(3)*pm.Normal('n0', mu=0, sd=1))
        x = pm.Deterministic('x',z)
        trace = pm.sample(2 * num_samples, chains=1)
        
        factual_z = trace['x']
        factual_x = trace['x']
        factual_y = trace['y']
        factual_n0 = trace['n0']

        counterfactual_x = 3 * np.ones(2 * num_samples)
        counterfactual_y = factual_y
        
        return factual_x, factual_y, counterfactual_x, counterfactual_y
```

And now we can sample from both the factual observed distribution $P(Y|X)$ and the counterfactual distribution $P(Y*|X^*=3)$.

```python
factual_x1, factual_y1, counterfactual_x1, counterfactual_y1 = model1_counterfactual_X_3()
factual_x2, factual_y2, counterfactual_x2, counterfactual_y2 = model2_counterfactual_X_3()
factual_x3, factual_y3, counterfactual_x3, counterfactual_y3 = model3_counterfactual_X_3()

plt.figure()
plt.title('Observational P(Y|X)')
kdeplot(factual_y1, color='blue', title='model1')
kdeplot(factual_y2, color='green', title='model2')
kdeplot(factual_y3, color='red', title='model3')
plt.legend()

plt.figure()
plt.title('Counterfactual P(Y* | X*=3)')
kdeplot(counterfactual_y1, color='blue', title='model1')
kdeplot(counterfactual_y2, color='green', title='model2')
kdeplot(counterfactual_y3, color='red', title='model3')
plt.legend()
```

Which confirms what we have shown in the previous results. I think the takeaway from this exercise was that you really have to think about the causal graph.
