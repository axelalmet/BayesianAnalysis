Here, we implement the Bayesian equivalent of the two group Chi-Squared Test, as demonstrated by Bob Carpenter in this blog post [here](https://lingpipe-blog.com/2009/10/13/bayesian-counterpart-to-fisher-exact-test-on-contingency-tables/).

```python
# Import the relevant libraries
import pymc3 as pm
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
def kdeplot(y,color,title):
    g = sns.kdeplot(y, color=color, label=title)
```

```python
num_samples = 1000 # Set the number of samples for inference algorithms
num_tune_steps = 10000 # Set the number of tuning steps

# Set the observed data values
K1 = 9
K2 = 4
N1 = 52
N2 = 48
```

```python
# Define the model
def model():
    with pm.Model():
        BoundedBeta = pm.Bound(pm.Beta, lower=0.0, upper=1.0)
        BoundedBinomial = pm.Bound(pm.Binomial, lower=0.0)
        
        # Define the priors
        theta1 = BoundedBeta('theta1', alpha=1 + K1, beta=1 + N1 - K1)
        theta2 = BoundedBeta('theta2', alpha=1 + K2, beta=1 + N2 - K2)
        
        # Defome the likelihoods
#         K1 = BoundedBinomial('K1', n=N1, p=theta1)
#         K2 = BoundedBinomial('K2', n=N2, p=theta2)
        
        # Sample from the posterior
        trace = pm.sample(num_samples, tune=num_tune_steps)
        
        return trace['theta1'], trace['theta2']
```

```python
theta1, theta2, K1, K2 = model()
```

```python
plt.figure()
plt.title('Posteriors for $\\theta_1$ and $\\theta_2$')
kdeplot(theta1, color='blue', title='$\\theta_1$')
kdeplot(theta2, color='red', title='$\\theta_2$')
plt.legend()
```

```python
plt.figure()
plt.title('Posteriors for $\\theta_2 - \\theta_1$')
kdeplot(theta2 - theta1, color='blue', title='$\\theta_2 - \\theta_1$')
```

```python
plt.figure()
plt.title('Posterioers for $K_1$ and $K_2$')
kdeplot(K1, color='blue', title='$K_1$')
kdeplot(K2, color='red', title='$K_2$')
plt.legend()
```
